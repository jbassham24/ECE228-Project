```python
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
import numpy as np
import os
```

```python
PATH_SOURCE = "/home/jbassham/jack/data/weddell/1992_2020"
PATH_DEST = os.path.join(PATH_SOURCE, 'plots_1')
```

```python
fnam = "stats_persistence_1992_2020.npz"

data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle = True)

pe_u_corr = data['corr_u']
pe_v_corr = data['corr_v']
pe_u_skill = data['skill_u']
pe_v_skill = data['skill_v']

lat = data['lat']
lon = data['lon']
```

```python
fnam = "stats_lr_1992_2020.npz"

data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle = True)

lr_u_corr = data['corr_u']
lr_v_corr = data['corr_v']
lr_u_skill = data['skill_u']
lr_v_skill = data['skill_v']
```

```python
fnam = "stats_cnn_1992_2020.npz"

data = np.load(os.path.join(PATH_SOURCE, fnam), allow_pickle = True)

cnn_u_corr = data['corr_u']
cnn_v_corr = data['corr_v']
cnn_u_skill = data['skill_u']
cnn_v_skill = data['skill_v']
```

```python
ilon_weddell = np.where((lon>= -70) & (lon<= 0))[0]
ilat_weddell = np.where((lat>=-80) & (lat<=-62))[0]
```

### LR - Persistence

```python
dcu = lr_u_corr - pe_u_corr
dcv = lr_v_corr - pe_v_corr
dsu = lr_u_skill - pe_u_skill
dsv = lr_v_skill - pe_v_skill

cmap = 'RdBu'
norm = TwoSlopeNorm(vcenter = 0)

# Plot difference (southern ocean)
fig, axs = plt.subplots(nrows = 4, ncols = 1, figsize = (12, 12))

plt0 = axs[0].pcolormesh(lon, lat, dcu, cmap = cmap, norm = norm)
fig.colorbar(plt0, ax = axs[0])
axs[0].set_title(r'$ \Delta $ corr u')
axs[0].set_xlabel('lon')
axs[0].set_ylabel('lat')

plt1 = axs[1].pcolormesh(lon, lat, dcv, cmap = cmap, norm = norm)
fig.colorbar(plt1, ax = axs[1])
axs[1].set_title(r'$ \Delta $ corr v')
axs[1].set_xlabel('lon')
axs[1].set_ylabel('lat')

plt2 = axs[2].pcolormesh(lon, lat, dsu, cmap = cmap, norm = norm)
fig.colorbar(plt2, ax = axs[2])
axs[2].set_title(r'$ \Delta $ skill u')
axs[2].set_xlabel('lon')
axs[2].set_ylabel('lat')

plt3 = axs[3].pcolormesh(lon, lat, dsv, cmap = cmap, norm = norm)
fig.colorbar(plt3, ax = axs[3])
axs[3].set_title(r'$ \Delta $ skill v')
axs[3].set_xlabel('lon')
axs[3].set_ylabel('lat')


fig.suptitle(f'LR - Persistence', fontweight = 'bold')
fig.tight_layout()

fnam = f"stats_dlrpe_1992_2020_wholeso.jpeg"
save_path = os.path.join(PATH_DEST, fnam)
plt.savefig(save_path, format='jpeg', dpi=300)

plt.show()
```
