```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_complex_yfit(fit, Y):
    """Plots real (u) and imaginary (v) componts of fit and y (true) values"""
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)

    # Plot
    fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(6, 10))
    
    axs[0].plot(t, uY, label = 'y')
    axs[0].plot(t, ufit, label = 'fit')
    axs[0].set_title(r'u components')
    axs[0].legend(loc='lower left')
    
    
    axs[1].plot(t, vY, label = 'y')
    axs[1].plot(t, vfit, label = 'fit')
    axs[1].set_title('v components')
    axs[1].legend(loc='lower left')
    
    fig.tight_layout()
    
    plt.show()
```

```python
def calc_r2(fit, Y):
    """Calculates r^2 for u, v components given true 'Y' and predicted 'fit' complex values"""
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)

    # u componenet
    ru = (ufit.flatten() - uY.T) # Residuals
    SSRu = np.sum(ru**2) # Sum of square residuals
    SSTu = np.sum((uY.T - np.mean(uY.T))**2) # Total sum of squares, mean should be zero
    r2u = (1-(SSRu/SSTu)) # R^2 value

    # v componenet
    rv = (vfit.flatten() - vY.T)
    SSRv = np.sum(rv**2)
    SSTv = np.sum((vY.T - np.mean(vY.T))**2) 
    r2v = (1-(SSRv/SSTv))

    return r2u, r2v
```

```python
def calc_skill(fit, Y):
    """Calculates model skill given true given true 'Y' and predicted 'fit' complex values"""
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)
    
    # Skill u component
    skill1u = np.sqrt(np.mean((ufit-uY)**2))
    skill2u = np.sqrt(np.mean((uY-np.mean(uY))**2))
    skillu = 1 - skill1u/skill2u

    # Skill v component
    skill1v = np.sqrt(np.mean((vfit-vY)**2))
    skill2v = np.sqrt(np.mean((vY-np.mean(vY))**2))
    skillv = 1 - skill1v/skill2v

    return skillu, skillv
```

```python
def calc_corr(fit, Y):
    """Calculates Pearson Correlation given true 'Y' and predicted 'fit' complex values"""
    # Phasing (no magnitude)
    # Decompose complex fit to u and v components
    ufit = np.real(fit)
    vfit = np.imag(fit)

    # Decompose complex Y values to u and v components
    uY = np.real(Y)
    vY = np.imag(Y)

    # Perason Correlation u
    corr1u = np.sum((uY-np.mean(uY))*(ufit-np.mean(ufit)))
    corr2u = np.sqrt(np.sum((uY-np.mean(uY))**2))*np.sqrt(np.sum((ufit-np.mean(ufit))**2))
    corru = corr1u/corr2u

    # Perason Correlation v
    corr1v = np.sum((vY-np.mean(vY))*(vfit-np.mean(vfit)))
    corr2v = np.sqrt(np.sum((vY-np.mean(vY))**2))*np.sqrt(np.sum((vfit-np.mean(vfit))**2))
    corrv = corr1v/corr2v

    return corru, corrv
```

```python
# Define set of sample points
t = np.arange(0, 2*2*np.pi, 0.1)
```

```python
nt = len(t)
print(nt)
```

```python
# Define noise/ residuals
noise = np.random.randn(nt)
```

***One Predictor Variable with No Noise***

```python
# Define amplitudes
c0 = 0 # Noise amplitude (starting at 0)
c1 = 1 # First parameter/ predictor variable amplitude

# Define frequencies k
k1 = 1

# Define true y values with noise amplitude c0
y = c0*noise + c1*np.cos(k1*t)

# Define E matrix predictor variables
E = np.array([np.cos(k1*t)]).T

# Initialize m matrix for coefficients
m = np.array([np.zeros(1)]).T

# Perform regression, no ridge (Wunsch EQN 2.95)
m = (np.linalg.inv(E.T@E))@E.T@y

# Calculate fit with regression coefficients
fit = E@m

print(f"shape E: {np.shape(E)}")
print(f"m: {m}")

# Plot final fit, y
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(6, 5))
axs.plot(t, y, label = 'y')
axs.plot(t, fit, label = 'fit')
axs.set_title(f'One Parameter, $c_0$ (noise) = {c0}')
axs.legend(loc='lower left')
plt.show()


# Calculate Statistics

r = (fit.flatten() - y.T) # Residuals
SSR = np.sum(r**2) # Sum of square residuals
SST = np.sum((y.T - np.mean(y.T))**2) # Total sum of squares, (* Not standardized, mean not 0 *)
r2 = (1-(SSR/SST)) # R^2 value
print (f"r^2 = {r2}")

# Pearson Correlation
corr1 = np.sum((y-np.mean(y))*(fit-np.mean(fit)))
corr2 = np.sqrt(np.sum((y-np.mean(y))**2))*np.sqrt(np.sum((fit-np.mean(fit))**2))
corr = corr1/corr2
print(f"Pearson Correlation = {corr}")

# Skill
skill1 = np.sqrt(np.mean((y-fit)**2))
skill2 = np.sqrt(np.mean((y-np.mean(y))**2))
skill = 1 - skill1/skill2
print(f"Skill = {skill}")
```

***Three Predictor Variables, Real, Same K, Ridge***

```python
# Create noise
c0 = 0 # Noise amplitude (starting at 0)
noise = c0*np.random.randn(nt)

# Create noise for input
c4 = 1
noise_in = c4*np.random.randn(nt)

# Define amplitudes for parameters
c1 = 1
c2 = 1
c3 = 1 
# Define frequencies for parameters
# If k1 = k2 or k1 = k3 or k2 = k3 than ridge needed (if columns E the same, they are not independent)
# Not independent gives singular matrix error (linearly dependent, no inverse)
k1 = 1
k2 = 1
k3 = 1

# Define true y values with noise amplitude c0
y = noise + c1*np.cos(k1*t) + c2*np.cos(k2*t) + c3*np.cos(k3*t)

# Define E matrix with predictor variables
E = np.array([noise_in + np.cos(k1*t), np.cos(k2*t), np.cos(k3*t)]).T
print(f"shape E: {np.shape(E)}")

# Create ridge parameter
I = np.identity(3)
lmdas = np.arange(10**-4, 1, 10**-3)

# Initialize r^2
nl = len(lmdas)
r2 = np.zeros(nl)

# Initialize dict for eignvalues, eigenvectors
eig_dict = {}

# Initialize counter
ct = 0

# Iterate through lamda values
for lmda in lmdas:
    # Calculate regression coefficiets with ridge parameter (Wunsch EQN 2.116)
    m1 = (np.linalg.inv(E.conj().T@E+I*lmda))@E.conj().T@y

    fit = E@m1

    # R^2
    r = (fit.flatten() - y.T) # Residuals
    SSR = np.sum(r**2) # Sum of square residuals
    SST = np.sum((y.T - np.mean(y.T))**2) # Total sum of squares, (* Not standardized, mean not 0 *)
    r2[ct] = (1-(SSR/SST)) # R^2 value

    # Eigenvalues and eigenvectors for each iteration in temp variables
    eigval_temp, eigvec_temp = np.linalg.eig(np.linalg.inv(E.conj().T@E+I*lmda))

    # Store eigenvalues and eigenvectors in dict
    eig_dict[ct] = {
        'eigvals': eigval_temp,
        'eigvecs': eigvec_temp
    }
    
    # Update counter
    ct+=1

print(f"r^2[0]: {r2[0]}, r^2[end]: {r2[-1]}")

# Plot final fit, y
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(6, 5))
axs.plot(t, y, label = 'y')
axs.plot(t, fit, label = 'fit')
axs.set_title(r'Three Parameters')
axs.legend(loc='lower left')
plt.show()

# Plot r^2
fig, axs = plt.subplots(nrows=1, ncols=1, figsize=(6, 5))
axs.plot(lmdas, r2)
axs.set_title(r'Ridge Regression')
axs.set_xlabel(xlabel = '$\lambda$', fontsize = 12)
axs.set_ylabel(ylabel = '$r^2$',  fontsize = 12)
plt.show()
```
